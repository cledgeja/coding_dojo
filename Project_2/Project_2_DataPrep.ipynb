# identify and addressed missing values 

```python
#missing values in this dataset were identified as '?'
#replace ? will nan to identify missing values
df.replace({'?':np.NaN},inplace=True)
```

```python
#missing values
df.isna().sum()[df.isna().sum()>0]
```

appears that all missing values are related to vector angles 

     Vector angles in degrees on front plane of:, linear
     10 QRS
     11 T
     12 P
     13 QRST
     14 J

```python
df[[10,11,12,13,14]].info()
```

The spatial QRS-T angle was categorized into three groups: normal (0 to 105°), borderline (105 to 135°) and abnormal (135 to 180°). The threshold for abnormal angles was based on a previous report. 23.
https://academic.oup.com/eurheartj/article/24/14/1357/501772


```python
df.groupby([13])[279].count()[df.groupby([13])[279].count()>1]
```

```python
df.replace({np.nan:0}).groupby([13])[279].count()[df.replace({np.nan:0}).groupby([13])[279].count()>1]
```

Appears that vector angles of 0 was not entered. can repace all missing with 0

```python
df.fillna(value=0,inplace=True);
```

```python
df.isna().sum()[df.isna().sum()>0]
```

```python
#make sure no object columns
dtypes = df.dtypes
str_cols = dtypes[dtypes =='object'].index
for col in str_cols:
  df[col] = df[col].astype('float64')
dtypes = df.dtypes
str_cols = dtypes[dtypes =='object'].index
str_cols
```
